# Handling Duplicates

This involves 2 steps: Detecting duplicates and Removing duplicates.

To check for the duplicates in our data

```python
df.duplicated()
```

All the function which is used in this file.

To remove the duplicates(if any) we can use the code :

```python

def clean_funding_round(domain_data):
    global df
    a=["series a", "venture a"]
    df = df[~df['SourceType'].isin(a)]
    return df.head()

def sort_domain_groupby(domain_data):
    domain_data = domain_data.sort_values(by= 'Original_Date', ascending=False)
    return domain_data.head()

def dedude(data):
    df=data.drop_duplicates()
    return df
```

```python
clean_funding_round(df)
```

```python
sort_domain_groupby(df)
```

```python
dedude(df)
```

# Missing Value Treatment

To treat the missing values we can opt for a method from the following :

1) Drop the variable

2) Drop the observation(s)

3) Missing Value Imputation



Since it is a categorical variable, letâ€™s impute the values by mode

```python
df.Domain.mode()[0]
```

```python
df.Domain.fillna(df.Domain.mode()[0],inplace=True)
```

```python
df.isna().sum()/df.shape[0]
```

```python
df.dropna(axis=1,inplace=True)
```
